PIVOT TABLES

```python
import pandas as pd
import numpy as np
```

```python
info = {'Company':['Ford', 'Ford', 'Vw', 'BMW', 'Cooper', 'Cooper'],
'Stars':[1, 2, np.nan, 2, 1, 1],
'Weight': [2, 4, 2, 2, 3, None],
'origin': ['China','Mexico', 'Mexico',None, 'China', pd.NaT],
'Length': [40, 50, 30, np.nan, 45, None]}
```

```python
type(info)
```

```python
df=pd.DataFrame(data=info)
df.head()
```

```python
df.isnull().any()
```

-------REMOVES ROWA  MISSING DATA--------

```python
ndf=df.copy()
ndf.dropna()
```

```python
ndf=df.copy()
ndf.dropna(axis=1)
```

```python
sm=df.Stars.mean()
```

```python
df
```

```python
df['Stars'].fillna(value = sm,
                   inplace = True)
df
```

```python
mw=df.Weight.mean()
mw
```

```python
df['Weight'].fillna(value= mw, inplace= True)
df
```

```python
ov=df.origin.mode()[1]
ov
```

```python
df['origin'].fillna(value=ov, inplace=True)
```

```python
df
```

```python
df.to_csv('test.csv')
```
